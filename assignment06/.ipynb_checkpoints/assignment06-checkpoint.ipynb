# import package for plotting, data management and dealing image

```python
import numpy as np
import matplotlib.pyplot as plt
```

# define data and make some noise

## x  : x-coordinate data

## y1 : (noisy) y-coordinate data

## y2 : (clean) y-coordinate data

```python
num     = 201
std     = 20
a       = 2 
b       = 10

n       = np.random.rand(num)
nn      = n - np.mean(n)
x       = np.linspace(-100,100,num)
y1      = a * x + nn * std + b 
y2      = a * x + b

plt.plot(x, y1, 'g.', x, y2, 'r--')
plt.show()
```

# find a line that fits the given data
### approximating line is obtained by the least square
### $y = f(x) = \theta1 * x + \theta2$
### $\theta = (A^TA)^{-1}A^Ty2$

```python
A=[np.ones(num)]
A.append(x)
A= np.array(A)
pinv_A=np.linalg.pinv(A)
theta=np.matmul(pinv_A.transpose(), y2)
print(theta)
```

# plot approximated line with data

```python
y3=np.matmul(A.transpose(), theta)
plt.plot(x, y1, 'y.', x, y3, 'b--')
plt.show()
```
