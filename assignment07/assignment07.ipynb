# import package for plotting, data management and dealing image

```python
import numpy as np
import matplotlib.pyplot as plt
```

# define data and make some noise

## x  : x-coordinate data

## y1 : (noisy) y-coordinate data

## y2 : (clean) y-coordinate data

```python
num     = 1001
std     = 5 

def fun(x):
 
	# f = np.sin(x) * (1 / (1 + np.exp(-x))) 
	f = np.abs(x) * np.sin(x)

	return f

n       = np.random.rand(num)
nn      = n - np.mean(n)
x       = np.linspace(-10,10,num)
y1      = fun(x)
y2      = y1 + nn * std

plt.plot(x, y1, 'b.', x, y2, 'k.')
plt.show()
```

# define function of finding a polynomial that fits the given data and calculate error
### approximating polynomial is obtained by the least square
### $y = f(x) = \theta1 * x + \theta2$
### $\theta = (A^TA)^{-1}A^Ty2$

```python
def PolyFit(p, x, y):
    A=[]
    for i in range(p+1):
        A.append(x**i)
    A= np.array(A)
    pinv_A=np.linalg.pinv(A)
    theta=np.matmul(pinv_A.transpose(), y)
    y_=np.matmul(A.transpose(), theta)
    d = (y_-y)**2
    distance= sum(d)
    plt.plot(x, y,'y.', x, y_,'b--')
    plt.title("%ith polynomial fitting" % p)
    plt.show()
    return distance
```

# plot approximated polynomial from 0 to 9

```python
d=np.zeros(10)
for i in range(10):
    d[i]=PolyFit(i, x, y2)
```

# show the error with x th approximated polynomial

```python
plt.plot(d)
plt.title("error with xth approximated polynomial")
plt.show()
```

# plot approximated polynomial from 10 to 19

```python
d1=np.zeros(10)
for i in range(10,20):
    d1[i-10]=PolyFit(i, x, y2)
```

# show the error with x th approximated polynomial

```python
x=np.linspace(10,20,10)
plt.plot(x, d1)
plt.title("error with xth approximated polynomial")
plt.show()
```
